```python
import numpy as np
import random
import pygame
```

```python
class Grid_world:

    def __init__(self, name = str, seed : int = 43):
        self.name = name
        self.rgn = random.Random(seed) 
        self.grid = np.ones((10,10), dtype=int)     
        self.rows, self.columns = self.grid.shape
        self.current_pos = [self.rgn.randrange(len(self.grid)), self.rgn.randrange(len(self.grid))]


    def generate_maze(self):
        print("random cell: ", self.current_pos)
        print("rows: ", self.rows, "columns: ", self.columns)
        path = []
            
        path.append(self.current_pos)
       

        while path:
            directions = [
                (self.current_pos[0]-2, self.current_pos[1]), (self.current_pos[0]+2,self.current_pos[1]),
                (self.current_pos[0], self.current_pos[1]-2), (self.current_pos[0],self.current_pos[1]+2)
            ]
            print("directions: ", directions)
            neighbors = []
            for n in directions:
                if  (n[0] >= 0 and n[0] < self.rows) and ( n[1]>= 0 and n[1]< self.columns):
                    if self.grid[n[0]][n[1]] == 1:
                        neighbors.append(n)
                        self.grid[n[0]][n[1]]= 0
                        self.grid[n[0]][n[1]]= 0
                        
                        
                    
```

```python
world = Grid_world(name = "maze")
world.grid
```
